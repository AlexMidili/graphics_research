```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
labels = ['Процесс "А"',
          'Процесс "Б"',
          'Процесс "В"',
          'Процесс "Г"',
          'Процесс "Д"',
          'ЕИКС']
values = [0.1,
          0.1,
          0.05,
          0.1,
          0.7,
          0.5]

explode = (0.05, 0.05, 0.05, 0.05, 0.1, 0.2)

fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels,
       shadow=True, explode=explode, rotatelabels=True)
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
# Линейная зависимость
x = np.linspace(0, 15, 30)
y1 = x
# Квадратичная зависимость
x1 = np.linspace(0, 50, 100)
y2 = [math.sqrt(i) for i in x]
# Построение графика
plt.figure(figsize=(12, 7))
plt.title("") # заголовок
plt.xlabel("Количество абонентов, единиц ")         # ось абсцисс
plt.ylabel("Загрузка сети, Мб")    # ось ординат
plt.plot(x, y1, x, y2)  # построение графика
plt.legend(['unicast', 'multicast'])
plt.show()
```
